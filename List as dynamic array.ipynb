## List as dynamic array
출처: https://www.youtube.com/watch?v=46j4IOxiN-w&list=PLXziV1DL41ojhfghjtP_jrU03iYy0vJZe&index=25

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# copy operation이 일어나는지 안 일어나는지 시간을 재어 알아보려 함

record_time = np.empty((1000000,)) # np.empty: 메모리를 잡아 놓고 그 안의 메모리는 건드리지 말라
temp = []
for i in range(1000000):
    tic = time.time()
    temp.append(i)
    toc=time.time()
    record_time[i] = toc-tic
    
plt.plot(record_time)
plt.savefig('Pictures/record_time.png')
plt.show()
plt.close('all')
```
